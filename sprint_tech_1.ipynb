```python
import pandas as pd
```

```python
data = pd.read_csv("base_datos_sti.csv")
data
```

```python
column = input("Ingresa la columna que deseas consultar: ")
orderInput = input("Deseas verlo en orden ascendente (1) o descendente (2): ")
def orderResult():
    if(orderInput == '1'):
        return True
    else:
        return False

data.sort_values(by=column, ascending=orderResult())
```
